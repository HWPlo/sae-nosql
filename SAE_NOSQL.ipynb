```python
# Question 5 :
# 5) Donner le nombre de commande pour chaque pays du client, ainsi que le montant total des commandes et le montant total payé : on veut conserver les clients n’ayant jamais commandé dans le résultat final ;

print("Question 5 :")
print("Donner le nombre de commande pour chaque pays du client, ainsi que le montant total des commandes et le montant total payé : on veut conserver les clients n’ayant jamais commandé dans le résultat final.")
print("")

q5 = db.customers.aggregate([
    
    # Désagréger pour avoir une ligne par transaction
    {
        "$unwind": {
            "path": "$Transac",
            "preserveNullAndEmptyArrays": True
        }
    },
    
    # Fusionner les commandes avec les clients
    {
        "$lookup": {
            "from": "orders",
            "localField": "customerNumber",
            "foreignField": "customerNumber",
            "as": "orders"
        }
    },
    
    # Désagréger pour avoir une ligne par commande
    {
        "$unwind": {
            "path": "$orders",
            "preserveNullAndEmptyArrays": True
        }
    },
    
    # Désagréger pour avoir une ligne par commande détaillée
    {
        "$unwind": {
            "path": "$orders.Comm",
            "preserveNullAndEmptyArrays": True
        }
    },
    
    # Garder les champs nécessaires pour les calculs et l'affichage
    {
        "$project": {
            "country": 1,
            "orderNumber": "$orders.orderNumber",
            "CommQuantity": {
                "$cond": {
                    "if": { "$isArray": "$orders.Comm.quantityOrdered" },
                    "then": { "$arrayElemAt": ["$orders.Comm.quantityOrdered", 0] },
                    "else": "$orders.Comm.quantityOrdered"
                }
            },
            "CommPrice": {
                "$cond": {
                    "if": { "$isArray": "$orders.Comm.priceEach" },
                    "then": { "$arrayElemAt": ["$orders.Comm.priceEach", 0] },
                    "else": "$orders.Comm.priceEach"
                }
            },
            "TransacAmount": {
                "$cond": {
                    "if": { "$isArray": "$Transac.amount" },
                    "then": { "$arrayElemAt": ["$Transac.amount", 0] },
                    "else": "$Transac.amount"
                }
            }
        }
    },
    
    # Ajouter les champs calculés pour les totaux des commandes et des transactions
    {
        "$addFields": {
            "CommTotal": {
                "$multiply": [
                    { "$toDouble": "$CommQuantity" },
                    { "$toDouble": "$CommPrice" }
                ]
            },
            "TransacAmount": {
                "$toDouble": "$TransacAmount"
            }
        }
    },
    
    # Grouper par pays pour calculer les totaux et listes uniques
    {
        "$group": {
            "_id": "$country",
            "numberOfOrders": { "$addToSet": "$orderNumber" },
            "totalOrderAmount": { "$sum": "$CommTotal" },
            "totalPaidAmount": { "$sum": "$TransacAmount" }
        }
    },
    
    # Calculs et formatage des champs
    {
        "$project": {
            "_id": 0,
            "country": "$_id",
            "numberOfOrders": { "$size": "$numberOfOrders" },
            "totalOrderAmount": {
                "$cond": {
                    "if": { "$eq": ["$totalOrderAmount", 0] },
                    "then": "nan",
                    "else": "$totalOrderAmount"
                }
            },
            "totalPaidAmount": {
                "$cond": {
                    "if": { "$eq": ["$totalPaidAmount", 0] },
                    "then": "nan",
                    "else": "$totalPaidAmount"
                }
            }
        }
    },
    
    # Trier les résultats par pays
    {
        "$sort": { 
            "country": 1 
        }
    }
])


pandas.DataFrame(list(q5))
```

```python
# Question 6 :
# 6) On veut la table de contigence du nombre de commande entre la ligne de produits et le pays du client ;

print("Question 6 :")
print("On veut la table de contigence du nombre de commande entre la ligne de produits et le pays du client.")
print("")

q6 = db.orders.aggregate([
    
    # Fusionner les clients avec les commandes pour récupérer les informations du pays
    {
        "$lookup": {
            "from": "customers",
            "localField": "customerNumber",
            "foreignField": "customerNumber",
            "as": "customer_info"
        }
    },
    
    # Extraire le pays du client dans un champ séparé
    {
        "$set": {
            "customerCountry": { "$arrayElemAt": ["$customer_info.country", 0] }
        }
    },
    
    # Désagréger les produits dans les commandes
    {
        "$unwind": {
            "path": "$Comm",
            "preserveNullAndEmptyArrays": True
        }
    },
    
    # Grouper par numéro de commande, ligne de produit et pays du client
    {
        "$group": {
            "_id": {
                "orderNumber": "$orderNumber",
                "productLine": "$Comm.productLine",
                "customerCountry": "$customerCountry"
            }
        }
    },
    
    # Grouper par ligne de produit et pays du client pour obtenir le nombre de commandes
    {
        "$group": {
            "_id": {
                "productLine": "$_id.productLine",
                "customerCountry": "$_id.customerCountry"
            },
            "orderCount": { "$sum": 1 }
        }
    },
    
    # Grouper par ligne de produit pour obtenir une liste des pays et le nombre de commandes par pays
    {
        "$group": {
            "_id": "$_id.productLine",
            "countries": {
                "$push": {
                    "country": "$_id.customerCountry",
                    "count": "$orderCount"
                }
            }
        }
    },
    
    # Désagréger la liste des pays et des commandes pour avoir une ligne par pays et ligne de produit
    {
        "$unwind": {
            "path": "$countries",
            "preserveNullAndEmptyArrays": True
        }
    },
    
    # Formater les résultats pour afficher la ligne de produit, le pays et le nombre de commandes
    {
        "$project": {
            "_id": 0,
            "productLine": "$_id",
            "country": "$countries.country",
            "numberOfOrders": "$countries.count"
        }
    },
    
    # Trier les résultats par ligne de produit puis par pays
    {
        "$sort": { 
            "productLine": 1, 
            "country": 1 
        }
    }
])


pandas.DataFrame(list(q6))
```

```python
# Question 7 :
# 7) On veut la même table croisant la ligne de produits et le pays du client, mais avec le montant total payé dans chaque cellule ;

print("Question 7 :")
print("On veut la même table croisant la ligne de produits et le pays du client, mais avec le montant total payé dans chaque cellule.")
print("")

q7 = db.orders.aggregate([
    
    # Relier les documents de la collection orders avec ceux de la collection customers en utilisant customerNumber 
    {
        "$lookup": {
            "from": "customers",  
            "localField": "customerNumber",  
            "foreignField": "customerNumber",  
            "as": "customer_info"  
        }
    },
    
    # Créer de nouveaux champs en extrayant des valeurs du tableau customer_info comme le pays du client et ses transactions depuis Transac
    {
        "$set": {
            "customerCountry": { "$arrayElemAt": ["$customer_info.country", 0] },  
            "customerTransactions": { "$arrayElemAt": ["$customer_info.Transac", 0] }  
        }
    },
    
    # Grouper les documents par numéro de commande, ligne de produit, pays et numéro de client
    {
        "$group": {
            "_id": { 
                "orderNumber": "$orderNumber",  
                "productLine": "$Comm.productLine",  
                "customerCountry": "$customerCountry",  
                "customerNumber": "$customerNumber"  
            },
            "customerTransactions": { "$first": "$customerTransactions" },  
            "Comm": { "$first": "$Comm" }  
        }
    },
    
    # Décomposer le tableau comm en plusieurs documents
    {
        "$unwind": "$Comm"
    },
    
    # Décomposer le tableau customerTransactions en plusieurs documents
    {
        "$unwind": "$customerTransactions"
    },
    
    # Regrouper à nouveau les documents en additionnant les montants des transactions pour chaque ligne de produit 
    {
        "$group": {
            "_id": {
                "orderNumber": "$_id.orderNumber",  
                "productLine": "$Comm.productLine",  
                "customerCountry": "$_id.customerCountry",  
                "customerNumber": "$_id.customerNumber"  
            },
            "totalAmount": { "$sum": "$customerTransactions.amount" }  # Somme des montants des transactions pour chaque groupe
        }
    },
    
    # Regrouper les résultats par ligne de produit et pays du client et additionner les montants totaux des commandes pour chaque pays
    {
        "$group": {
            "_id": {
                "productLine": "$_id.productLine", 
                "customerCountry": "$_id.customerCountry"  
            },
            "totalPaidAmount": { "$sum": "$totalAmount" }  
        }
    },
    
    # Affichage final 
    {
        "$project": {
            "_id": 0,  
            "productLine": "$_id.productLine",  
            "country": "$_id.customerCountry",  
            "totalPaidAmount": "$totalPaidAmount"  
        }
    },
    
    # Trier les résultats par ligne de produit et pays
    {
        "$sort": {
            "productLine": 1,  
            "country": 1  
        }
    }
]) 


pandas.DataFrame(list(q7))
```

```python
# Question 8 :
# 8) Donner les 10 produits pour lesquels la marge moyenne est la plus importante ;

print("Question 8 :")
print("Donner les 10 produits pour lesquels la marge moyenne est la plus importante.")
print("")

q8 = db.orders.aggregate([
    
    # Décomposer chaque élément du tableau comm
    {
        "$unwind": "$Comm"
    },
    
    # Sélectionner les champs que nous voulons garder + ajouter la marge (prix de vente - prix d'achat)
    {
        "$project": {
            "productCode": "$Comm.productCode", 
            "productName": "$Comm.productName",
            "margin": { "$subtract": ["$Comm.priceEach", "$Comm.buyPrice"] }
        }
    },
    
    # Grouper les résultats par code produit et nom de produit et calculer la marge moyenne pour chaque produit
    {
        "$group": {
            "_id": {
                "productCode": "$productCode",
                "productName": "$productName"
            },
            "averageMargin": { "$avg": "$margin" }
        }
    },
    
    # Affichage final 
    {
        "$project": {
            "_id": 0,
            "productCode": "$_id.productCode",
            "productName": "$_id.productName",
            "averageMargin": "$averageMargin"
        }
    },
    
    # Trier en fonction de la marge 
    {
        "$sort": { "averageMargin": -1 }
    },
    
    # TOP10 
    {
        "$limit": 10
    }
])


pandas.DataFrame(list(q8))
```

```python
# Question 9 :
# 9) Lister les produits (avec le nom et le code du client) qui ont été vendus à perte :
# Si un produit a été dans cette situation plusieurs fois, il doit apparaître plusieurs fois,
# Une vente à perte arrive quand le prix de vente est inférieur au prix d’achat ;

print("Question 9 :")
print("Lister les produits (avec le nom et le code du client) qui ont été vendus à perte :")
print("Si un produit a été dans cette situation plusieurs fois, il doit apparaître plusieurs fois,")
print("Une vente à perte arrive quand le prix de vente est inférieur au prix d’achat.")
print("")

q9 = db.orders.aggregate([
    
    # Décomposer chaque élément du tableau comm
    {
        "$unwind": "$Comm"
    },
    
    # Ajouter un champ Perte pour déterminer si le produit a été vendu à perte
    {
        "$addFields": {
            "Perte": {"$lt": ["$Comm.priceEach", "$Comm.buyPrice"]}
        }
    },
    
    # Filtrer les documents pour ne garder que les ventes à perte
    {
        "$match": {"Perte": True}
    },
    
    # Effectuer une jointure avec la collection customers pour obtenir les informations sur le client
    {
        "$lookup": {
            "from": "customers",
            "localField": "customerNumber",
            "foreignField": "customerNumber",
            "as": "customerInfo"
        }
    },
    
    # Regrouper les résultats par code produit, numéro client, prix de vente et prix d'achat
    {
        "$group": {
            "_id": {
                "productCode": "$Comm.productCode",
                "customerNumber": "$customerNumber",
                "priceEach": "$Comm.priceEach",
                "buyPrice": "$Comm.buyPrice"
            },
            "productName": {"$first": "$Comm.productName"},
            "customerName": {"$first": {"$arrayElemAt": ["$customerInfo.customerName", 0]}}
        }
    },
    
    # Trier les résultats en fonction du numéro client
    {
        "$sort": {"_id.customerNumber": -1}
    },
    
    # Affichage final 
    {
        "$project": {
            "_id": 0,
            "productCode": "$_id.productCode",
            "productName": "$productName",
            "customerName": "$customerName",
            "customerNumber": "$_id.customerNumber",
            "priceEach": "$_id.priceEach",
            "buyPrice": "$_id.buyPrice"
        }
    }
])


pandas.DataFrame(list(q9))
```
